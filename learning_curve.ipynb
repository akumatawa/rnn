```python
# %matplotlib inline
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/drive')
```

```python
training = []
validation = []

with open(f'/drive/My Drive/science/125_125.txt', 'r') as f:
    for line in f.readlines():
        t, v = line.split(' ')
        training.append(float(t))
        validation.append(float(v))

print(len(training))
print(len(validation))
```

```python
drive.flush_and_unmount()
```

```python
plt.style.use('seaborn')

epoches = [i for i in range(1, 21)]

plt.plot(epoches, training, label='train', linewidth=2.0)
plt.plot(epoches, validation, label='validation', linewidth=2.0)

plt.legend()

plt.show()
```
